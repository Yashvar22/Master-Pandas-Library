```python
import numpy as np;
import pandas as pd;
```

Find Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Remove null values

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3) # rows mai utni non null value honi chahiye
```

Filling the missing data


```python
df.fillna(0)
```

```python
values={'A':0,'B':1,'C':300,'D':400}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
